```python
def parseOutSqm(x):
    return x[:len(x)-6]

def formattingGroundFloor(f):
    # Parsing out the sq. m
    f["GroundFloorArea"] = f["GroundFloorArea"].apply(parseOutSqm).astype(float)

    f["GroundFloorArea"] = f["GroundFloorArea"].fillna(f["GroundFloorArea"].median())

formattingGroundFloor(training)
formattingGroundFloor(final)
```

```python
def parseYear(x):
    return x[1:]
def formattingYear(f):
    f["Year"] = f["Year"].apply(parseYear)

formattingYear(training)
formattingYear(final)
```

```python
def formattingTypes(f):
    # Converting types to numbers
    f.loc[f["Type"] == "House", "Type"] = 0
    f.loc[f["Type"] == "Apartment", "Type"] = 1
    f.loc[f["Type"] == "Ground-floor apartment", "Type"] = 2
    f.loc[f["Type"] == "Mid-floor apartment", "Type"] = 3
    f.loc[f["Type"] == "Top-floor apartment", "Type"] = 4
    f.loc[f["Type"] == "End of terrace house", "Type"] = 5
    f.loc[f["Type"] == "Mid-terrace house", "Type"] = 6
    f.loc[f["Type"] == "Semi-detached house", "Type"] = 7
    f.loc[f["Type"] == "Detached house", "Type"] = 8
    f.loc[f["Type"] == "Maisonette", "Type"] = 9
    f["Type"]=pd.to_numeric(f["Type"], errors=0)

formattingTypes(training)
formattingTypes(final)
```

```python
def formattingClasses(f): 
    # Converting classes to numbers
    # 
    f.loc[f["EnergyRatingCat"] == "A1", "EnergyRatingCat"] = 0
    f.loc[f["EnergyRatingCat"] == "A2", "EnergyRatingCat"] = 1
    f.loc[f["EnergyRatingCat"] == "A3", "EnergyRatingCat"] = 2
    f.loc[f["EnergyRatingCat"] == "B1", "EnergyRatingCat"] = 3
    f.loc[f["EnergyRatingCat"] == "B2", "EnergyRatingCat"] = 4
    f.loc[f["EnergyRatingCat"] == "B3", "EnergyRatingCat"] = 5
    f.loc[f["EnergyRatingCat"] == "C1", "EnergyRatingCat"] = 6
    f.loc[f["EnergyRatingCat"] == "C2", "EnergyRatingCat"] = 7
    f.loc[f["EnergyRatingCat"] == "C3", "EnergyRatingCat"] = 8
    f.loc[f["EnergyRatingCat"] == "D1", "EnergyRatingCat"] = 9
    f.loc[f["EnergyRatingCat"] == "D2", "EnergyRatingCat"] = 10
    f.loc[f["EnergyRatingCat"] == "E1", "EnergyRatingCat"] = 11
    f.loc[f["EnergyRatingCat"] == "E2", "EnergyRatingCat"] = 12
    f.loc[f["EnergyRatingCat"] == "F ", "EnergyRatingCat"] = 13
    f.loc[f["EnergyRatingCat"] == "G ", "EnergyRatingCat"] = 14
    
formattingClasses(training)
```

```python
def getInterestingFields(f):
    # Selecting the data we want
    usableData=f[['BuildingID','Type','GroundFloorArea', 'ExposedRoofArea','SecSHFuel', 'AvgWallU', 'AvgWindowU', 'ExposedWallArea','AvgRoofU', 'SecSHFrac']]
    usableData['AvgWallU']= pd.to_numeric(usableData['AvgWallU'])
    usableData['AvgWindowU']= pd.to_numeric(usableData['AvgWindowU'])
    usableData['AvgRoofU']= pd.to_numeric(usableData['AvgRoofU'])
    usableData['SecSHFrac']= pd.to_numeric(usableData['SecSHFrac'])
    usableData['SecSHFuel']= pd.to_numeric(usableData['SecSHFuel'], errors=0)    
    return usableData

usableData = getInterestingFields(training)
finalData = getInterestingFields(final)
usableData.head(10)
```
